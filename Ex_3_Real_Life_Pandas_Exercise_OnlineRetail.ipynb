### 1. Load the dataset and display the first 5 rows.
üìå Use `pd.read_excel()` and parse dates. #Try parse_dates arg from pd.read_excel() function

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
retail=pd.read_excel('/content/Online Retail.xlsx',parse_dates=['InvoiceDate'])
retail.head()
```

### 2. Create a new column `TotalPrice` = `Quantity` √ó `UnitPrice` using `.apply()`.

```python
retail['TotalPrice']= retail.apply(lambda x: x['Quantity']*x['UnitPrice'], axis=1)
retail.head(2)
```

### 3. Filter all transactions where `TotalPrice` is above ¬£1000.

```python
retail[retail['TotalPrice']>1000]
```

### 4. Sort the dataset by `TotalPrice` in descending order.

```python
retail[retail['TotalPrice']>1000].sort_values('TotalPrice',ascending=False)[['InvoiceNo','TotalPrice']].head()
```

### 5. Group the data by `Country` and compute total `Quantity` and `TotalPrice`.

```python
retail.groupby('Country')[['Quantity','TotalPrice']].sum().sort_values('TotalPrice',ascending=False).head()
```

### 6. Find the top 5 customers by total spend.

```python
retail.groupby('CustomerID')[['TotalPrice']].sum().sort_values('TotalPrice',ascending=False).head()
```

### 7. Extract the month from `InvoiceDate` and count transactions per month.

```python
retail['InvoiceDate'].dt.month.value_counts().sort_values(ascending=False)
```

### 8. Filter transactions from the UK with positive `Quantity`.

```python
retail[(retail['Country']=='United Kingdom') & (retail['Quantity']>0) ].head(2)
```

### 9. Create a column `ValueLabel` as 'High' if `TotalPrice` > 500 else 'Low'.

```python
retail['ValueLabel']=np.where(retail['TotalPrice']>500,'High','Low')
retail[['TotalPrice','ValueLabel']].head()
```

### 10. Count number of `High` value transactions by country.

```python
retail.groupby('Country')[['ValueLabel']].value_counts().head()
```
